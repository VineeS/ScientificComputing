# **Introduction**

##System of nonlinear equation 

There are mainly two kinds of equation linear and nonlinear. The linear equation or a function can be represented in the following form 

$Ax + Bx = 0$ --------- **Eq(1)**,  or 

$f(x) = a_1 x_1+a_2 x_2+a_3 x_3……+a_n x_n - b = 0$ ------ **Eq(2)**

where $a’s$ and b are constants [2]. A non-linear equation can be $a_1x^2+a_2y^2 = b$ i.e. It can be defined as the equation which cannot be written in the form of **Eq(1)** or **Eq(2)**.

Now we are given the following equations:
1. Planet CISC-600 has an elliptical orbit that can be described in a plane by the equation

>$p_1 x^2 + p_2 x + p_3 y^2+p_4 y = p_5$, which can be written as

>$p_1 x^2 + p_2 x + p_3 y^2+p_4 y - p_5 = 0 **------ Eq(3)**

2. Comet HU has an elliptical orbit that can be described in the same plane by the equation

>$c_1 x^2 + c_2 x + c_3 y^2+c_4 y = c_5$,  which can be written as

>$c_1 x^2 + c_2 x + c_3 y^2+c_4 y - c_5 = 0 $ **----- Eq(4)**

The equation **Eq(3)** and **Eq(4)** are nonlinear equation as it cannot be written in the form of  **Eq(1)** or **Eq(2)**.


##Newton Raphson Method
It can be defined as quick way to find an approximation of the root of a real valued function, It uses the idea that a continuous and differentiable function can be approximated by a straight line tangent to it. **Refer Fig 1**, if the initial guess at the root is $x_i$, a tangent can be extended from the point $[x_i, f(x_i)]$. The point where this tangent crosses the x axis usually represents an improved estimate of the root. [3]

![Figure 1](https://drive.google.com/uc?id=1_-KfZoT8kIxuGMyN9GRoYXxqA_ttMvGU)
**Fig 1**

The Newton Raphson method can be derived from the above geometrical figure as,
 

>$f’(x_i) = \frac{f(x_i-0)}{x_i - x_{i+1}}$

>$x_{i+1} = x_i  - \frac{f(x_i)}{f’(x_i)}$ **--- Newtorn Raphson Eq(5)**

The above formula works i.e. Eq(5) for linear equations i.e. equations in form of 

>$a_1 x+a_2 y = 0$

>$a_3 x-a_4 y = 0$

The Eq(3) and Eq(4) are Nonlinear Algebraic equation hence, we will be using the System of Non-linear Equation which is generalised as $f(x)=0$, [4]

Where $x$ and $f$ are $n$-vectors 

>$x =
 \begin{pmatrix}
  x_1\\
  x_2 \\
  \vdots\\
  x_n
  \end{pmatrix}$  $f(x) =
 \begin{pmatrix}
  f_1(x_1,x_2,& \cdots & x_n)\\
  f_2(x_1,x_2,& \cdots & x_n) \\
  \vdots  & \vdots   & \vdots  \\
  f_n(x_1,x_2,& \cdots & x_n)
  \end{pmatrix}$

The Newton Method vector version will be [ from Eq(5) ]  $x_{i+1} = x_i  - f(x_i)$, the problem comes when we have do division by first order derivative of $f(x_i)$ i.e. $f’(x_i)$. The concept of **division is not allowed in vectors or matrices** hence, the division will be replaced with the inverse of $f'(x_i)$. For which the first step will be to check for the vectors containing derivative of  $f(x_i)$.**The Jocobian is basically a method that represents the equivalent of a derivative that we have in Newton’s method**. The ***Jocobian*** method will help finding the inverse. It can be represented in the following way [3]

> $J  =
 \begin{pmatrix}
  \frac{\partial f_1}{\partial x_1},\frac{\partial f_1}{\partial x_2},& \cdots & \frac{\partial f_1}{\partial x_n}\\
  \frac{\partial f_2}{\partial x_1},\frac{\partial f_2}{\partial x_2},& \cdots & \frac{\partial f_2}{\partial x_n} \\
  \vdots  & \vdots   & \vdots  \\
  \frac{\partial f_n}{\partial x_1},\frac{\partial f_n}{\partial x_2},& \cdots & \frac{\partial f_n}{\partial x_n}
  \end{pmatrix}$

The Jocobian is basically a method that represents the equivalent of a derivative that we have in Newton’s method, $J$ is a matrix instead of dividing by $f’(x)$ we multiply by $J^{-1}$. Substitute Eq(5)  $f’(x)$  by $J^{-1}(x_i)$ as follows:

> $x_{i+1} = x_i  - f(x_i)  J^{-1}(x_i)$ **-- Eq(6)**


Solving an inverse is very **Eq(6)** expensive hence, we can rearrange the equation 

> $J(x_i)\partial x_i  = -f(x_i)$  **---Eq(7)** where $\partial x_i = x_{i+1}-x_i$ **---Eq(8)**

This Newton's method for solving nonlinear systems of Algebraic equations will be used to get the possible point of collision. Which will be explained in detail the method section of the document.

```
import numpy as np
import matplotlib.pyplot as plt
def two_orbits_plot(p1, p2, p3, p4, p5,
                  c1, c2, c3, c4, c5,
                  xhi=5, xlo=-5, yhi=5, ylo=-5):
    
    xx=np.linspace(xlo,xhi)
    yy=np.linspace(ylo,yhi)
    [X,Y]=np.meshgrid(xx,yy)
    P = p1*X**2+p2*X+p3*Y**2+p4*Y**2-p5
    C = c1*X**2+c2*X+c3*Y**2+c4*Y-c5
    
    
    
    plt.figure()
    plt.contour(X,Y,P,[1])
    plt.contour(X,Y,C,[1], linestyles= '--')
    plt.title("Comet(Dashed) and Planet(Solid) orbits")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    

print("                   Plot 1                             ")
two_orbits_plot(p1=16.0, p2=44.0, p3=4.0, p4=0.0, p5=64.0,
                  c1=0.8, c2=-4.0, c3=27.0, c4=-22.0, c5=6.0,
                  xhi=5, xlo=-5, yhi=5, ylo=-5)

print('\n----------------------------------------------------------\n')

def Jacobian_matrix(xy):
    X, Y = xy # taken xy as an array
    p1=16.0
    p2=44.0
    p3=4.0
    p4=0.0
    p5=64.0
    c1=0.8
    c2=-4.0
    c3=27.0
    c4=-22.0
    c5=6.0

    return [[2*X*p1+p2,2*Y*p3+p4],[2*X*c1+c2,2*Y*c3+c4]]



def functions_matrix(xy):
    X, Y = xy # taken xy as an array
    p1=16.0
    p2=44.0
    p3=4.0
    p4=0.0
    p5=64.0
    c1=0.8
    c2=-4.0
    c3=27.0
    c4=-22.0
    c5=6.0
    return [p1*X**2+p2*X+p3*Y**2+p4*Y-p5, c1*X**2+c2*X+c3*Y**2+c4*Y-c5]



def newton_raphson_iterative(function, x_init, jacobian):
    max_iteration = 50
    epsilon = 1e-6
    print("Newton Rapson Iteration ")
    print(" x       ,        y  ")
    x_last = x_init
    for k in range(max_iteration):
        J = np.array(Jacobian_matrix(x_last)) #... Step to find Jacobian matrix
        F = np.array(functions_matrix(x_last)) #... Step to find Function matrix

        #using canned function to get the value of del_x
        del_x = np.linalg.solve( J, -F )
        
        #Calculating x_i+1, new value of x
        x_new = x_last + del_x
        print(x_last)

        diff_xl_xn = np.linalg.norm(x_last - x_new)
        #checking if value of x_i converges
        if diff_xl_xn < epsilon:
            print('converged in iteration:', k )
            break
        else:
            #Reassign x_i to x_i+1 and repeat
            x_last= x_new

    else: # only if the for loop end 'naturally'
        print('not converged')

    return x_last



root1 = newton_raphson_iterative(functions_matrix, [-1.0,-2.0], Jacobian_matrix)
#print('solution example:', root1 )
x =functions_matrix(root1)  
print("f(xroot,yroot):",x[0])
print("g(xroot,yroot):",x[1])
print("----------------------------------------------------------------")
root2 = newton_raphson_iterative(functions_matrix, [1.0,1.0], Jacobian_matrix)
#print('solution example:', root2 )
y = functions_matrix(root2) 
print("f(xroot,yroot):",y[0])
print("g(xroot,yroot):",y[1])

```
