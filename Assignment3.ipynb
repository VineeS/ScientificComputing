Firstly let's understand Root Finding Method, it is mainly an algorithm for finding ‘zeroes’ also called as roots of a function [1]. For example, roots of a quadratic equation [4]- 

>$f(x) = ax^2+bx+c = 0$ -------------- $Eq(1)$

can be calculated as below

> $x = \frac{-b \pm \sqrt{b^2-4ac}}{2a}$  ---------- $Eq(2)$

$Eq(2)$ - Represents the value of **x** which makes the $Eq(1)$ zero. Thus, we can say **x** is the root of an equation as it makes $f(x) = 0$. 

For a simple quadratic function like in $Eq(1)$, we were able to solve using a formula but there are other functions for which roots cannot be calculated so easily. Hence for these functions other roots finding methods are used, which divided into two sub-groups i.e. bracketing methods and open methods which will be mainly explained by this document. 

##Bisection Method

Bisection method is a type of bracketing method. 

Bracketing methods determine continuously smaller brackets (intervals) that contain a root by exploiting a fact that the function changes it’s sign in the locality of the root. When the intervals become smaller it can be said that a root is been found. $[1]$ It requires two initial guesses of the root that must be on either side of the root.

The simplest root-finding method is bisection method. This is a type of incremental search method,  the intervals i.e. the upper limit and lower limit is divided by two. If a function changes sign over an interval, the function value at the midpoint is evaluated. The location of the root is then determined as lying at the midpoint of the subinterval within which the sign change occurs. The process is repeated until a refined approximation is obtained. The pseudo code can be as follows:  $[4]$

**Pseudo code 1**
1. Let’s consider upper limit as $b$ and lower limit as $a$. Now calculating the function value for $a$ and $b$ and taking a root such that function changes the sign over the interval i.e. $f(a)*f(b) < 0$
2. Now calculating the root i.e. 
	$c = \frac{a+b}{2}$
3.   Function value at the midpoint i.e. $f(c)$
4.   Now check which sub-intervals the root lies-
      *  If  $f(a)*f(c) < 0$ , the root is in lower interval. Hence substitute b = c and 		function $f(b) = f(c)$. Then return to step 2.
      *	 If  $f(a)*f(c) > 0$ , the root is in upper interval. Hence substitute a = c and 		function $f(a) = f(c)$. Then return to step 2.

      *	 If $f(a)f(c) == 0$, the root equals $c$ and process terminates.


##Secant Method

Secant method is a type of open method. Open methods are based on formulas unlike bracketing method. These formulas require only one starting value say **x** or sometimes two starting values that do not bracket the root. As the computation progress they sometime tends to diverge or move away from the root. However, they are faster than bracketing methods when converges[4]. Secant methods are little similar to bisection method, except it rather divides each interval by the secant line connecting the endpoints. The pseudo code for secant method is as followed -

**Pseudo code 2**
1. Let’s consider lower limit as $a$ and upper limit as $b$. Now calculating the function value for $a$ and $b$ and taking a root such that function changes the sign over the interval i.e. $f(a)*f(b) < 0$.
2. Now calculating the secant line for which secant formula is used as follows -
   $c = a - f(a)*\frac{b-a}{f(b)-f(a)}$
3. Now check which subinterval does the root lies -
     * If $f(a)==f(b)==f(c) or f(c) == 0 , root value found and process terminate.

     * If  $f(a)*f(c) < 0$ , then let [a, b] the next interval with a= a and b = c return 		to step 2

     * If  $f(b)*f(c) < 0$ , then let [a, b] the next interval with a = c and b= b return 		to step 2

###Finding the Mach 1 time of a launching rocket

Our aim is to estimate Mach 1 time of a launching rocket, Mach 1 is the ratio of the speed of an object to the speed of sound. When the rocket is flying less than Mach 1 it's in subsonic speed as soon as it reaches to the speed of sound we can say that the rocket has reached Mach 1 speed. For calculating the time at which the Andromeda 6 rocket reaches the speed of sound in a vertical flight around the surface of a planet i.e. Mach 1 time can be approximated by using the following formula:

> $ v = u * ln \frac {m_0}{m_0-dmdt*t}-g*t $. ----- $Eq(3)$

And substituting $v_s$ (speed of sound) in $v$ and rearranging the equation.

> $u * ln \frac {m_0}{m_0-dmdt*t}-g*t - vs = 0 $. ----- $Eq(4)$

Once we have our equation the next step is to substitute the values in the above equation,  given :

$u$: exhaust velocity, $m_0$: rocket mass, $dmdt$: rate of fuel consumption, $g$: gravitational acceleration, $v_s$: speed of sound, $a$: left interval endpoint, $b$: right interval endpoint and epsilon: tolerance, to calculate Mach 1 time for the rocket launched from Neptune.

This ***t*** determines the time at which the rocket has reached Mach 1 speed.

3. The third function is vs_time_bisect, in this method I have used the bisection method to calculate the value of ***t***. The function follows the pseudo code mentioned above in the introduction part (**Pseudo code 1**).
